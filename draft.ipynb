```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from plotnine import *
import seaborn as sns
from datetime import datetime, timedelta
import numpy as np

# ==============================================================================
# OPCIÓN 1: PLOTNINE (ggplot style) - RECOMENDADA PARA QUARTO
# ==============================================================================


def create_timeline_plotnine(timeline_dict):
    """
    Create a timeline plot using plotnine (ggplot2 style)
    Best for Quarto integration and static visualization
    """
    # Convert dict to DataFrame
    df = pd.DataFrame(
        [
            {"event": event, "date": date, "category": info.get("category", "Default")}
            for event, info in timeline_dict.items()
            for date in (
                info["date"] if isinstance(info["date"], list) else [info["date"]]
            )
        ]
    )

    # Ensure date column is datetime
    df["date"] = pd.to_datetime(df["date"])

    # Create the plot
    plot = (
        ggplot(df, aes(x="date", y="event", color="category"))
        + geom_point(size=4, alpha=0.8)
        + geom_segment(
            aes(x="date", xend="date", y="event", yend="event"), size=2, alpha=0.6
        )
        + theme_minimal()
        + theme(
            axis_text_x=element_text(rotation=45, hjust=1),
            panel_grid_major_y=element_line(color="lightgray", alpha=0.5),
            panel_grid_minor=element_blank(),
            legend_position="bottom",
        )
        + labs(title="Timeline de Eventos", x="Fecha", y="Eventos", color="Categoría")
        + scale_x_datetime(date_labels="%Y-%m-%d", date_breaks="1 month")
    )

    return plot


# ==============================================================================
# OPCIÓN 2: MATPLOTLIB - MÁS CONTROL Y CUSTOMIZACIÓN
# ==============================================================================


def create_timeline_matplotlib(timeline_dict):
    """
    Create timeline using matplotlib - maximum customization
    """
    # Prepare data
    events = []
    dates = []
    categories = []

    for event, info in timeline_dict.items():
        event_dates = info["date"] if isinstance(info["date"], list) else [info["date"]]
        for date in event_dates:
            events.append(event)
            dates.append(pd.to_datetime(date))
            categories.append(info.get("category", "Default"))

    df = pd.DataFrame({"event": events, "date": dates, "category": categories})

    # Create figure
    fig, ax = plt.subplots(figsize=(12, 8))

    # Color mapping
    unique_categories = df["category"].unique()
    colors = plt.cm.Set3(np.linspace(0, 1, len(unique_categories)))
    color_map = dict(zip(unique_categories, colors))

    # Plot events
    for i, (event, group) in enumerate(df.groupby("event")):
        y_pos = i
        for _, row in group.iterrows():
            color = color_map[row["category"]]
            ax.scatter(row["date"], y_pos, c=[color], s=100, alpha=0.8, zorder=3)
            ax.plot(
                [row["date"], row["date"]],
                [y_pos - 0.1, y_pos + 0.1],
                color=color,
                linewidth=3,
                alpha=0.7,
            )

    # Customize
    ax.set_yticks(range(len(df["event"].unique())))
    ax.set_yticklabels(df["event"].unique())
    ax.grid(True, alpha=0.3)
    ax.set_xlabel("Fecha")
    ax.set_ylabel("Eventos")
    ax.set_title("Timeline de Eventos")

    # Format dates
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    plt.xticks(rotation=45)

    # Legend
    legend_elements = [
        plt.Line2D(
            [0],
            [0],
            marker="o",
            color="w",
            markerfacecolor=color_map[cat],
            markersize=10,
            label=cat,
        )
        for cat in unique_categories
    ]
    ax.legend(handles=legend_elements, loc="upper right")

    plt.tight_layout()
    return fig


# ==============================================================================
# OPCIÓN 3: SEABORN - ELEGANTE Y SIMPLE
# ==============================================================================


def create_timeline_seaborn(timeline_dict):
    """
    Create timeline using seaborn - clean and elegant
    """
    # Prepare data
    data = []
    for event, info in timeline_dict.items():
        event_dates = info["date"] if isinstance(info["date"], list) else [info["date"]]
        for date in event_dates:
            data.append(
                {
                    "event": event,
                    "date": pd.to_datetime(date),
                    "category": info.get("category", "Default"),
                }
            )

    df = pd.DataFrame(data)

    # Create plot
    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=df, x="date", y="event", hue="category", s=150, alpha=0.8)

    # Add vertical lines
    for _, row in df.iterrows():
        plt.axvline(x=row["date"], alpha=0.3, linestyle="--", linewidth=1)

    plt.xticks(rotation=45)
    plt.title("Timeline de Eventos")
    plt.xlabel("Fecha")
    plt.ylabel("Eventos")
    plt.grid(True, alpha=0.3)
    plt.tight_layout()

    return plt.gcf()


# ==============================================================================
# OPCIÓN 4: GANTT-STYLE TIMELINE (PARA RANGOS DE FECHAS)
# ==============================================================================


def create_gantt_timeline(timeline_dict):
    """
    Create Gantt-style timeline for date ranges
    """
    # Prepare data for ranges
    data = []
    for event, info in timeline_dict.items():
        if isinstance(info["date"], dict) and "start" in info["date"]:
            # Range format
            start_date = pd.to_datetime(info["date"]["start"])
            end_date = pd.to_datetime(info["date"].get("end", info["date"]["start"]))
        else:
            # Single date - create small range
            date = pd.to_datetime(info["date"])
            start_date = date
            end_date = date + timedelta(days=1)

        data.append(
            {
                "event": event,
                "start": start_date,
                "end": end_date,
                "duration": (end_date - start_date).days,
                "category": info.get("category", "Default"),
            }
        )

    df = pd.DataFrame(data)

    # Create Gantt chart with plotnine
    plot = (
        ggplot(df, aes(y="event"))
        + geom_segment(
            aes(x="start", xend="end", yend="event", color="category"),
            size=6,
            alpha=0.7,
        )
        + geom_point(aes(x="start", color="category"), size=3)
        + geom_point(aes(x="end", color="category"), size=3)
        + theme_minimal()
        + theme(
            axis_text_x=element_text(rotation=45, hjust=1),
            panel_grid_major_y=element_line(color="lightgray", alpha=0.5),
        )
        + labs(
            title="Timeline Gantt de Eventos", x="Fecha", y="Eventos", color="Categoría"
        )
        + scale_x_datetime(date_labels="%Y-%m-%d")
    )

    return plot


# ==============================================================================
# EJEMPLO DE USO
# ==============================================================================

# Timeline data example
timeline_data = {
    "Evento 1": {"date": "2024-01-15", "category": "Inicio"},
    "Evento 2": {"date": "2024-02-20", "category": "Desarrollo"},
    "Evento 3": {"date": "2024-03-10", "category": "Desarrollo"},
    "Evento 4": {"date": "2024-04-05", "category": "Finalización"},
    "Evento 5": {
        "date": ["2024-05-01", "2024-05-15"],  # Multiple dates
        "category": "Review",
    },
}

# Gantt timeline data example (with ranges)
gantt_data = {
    "Fase 1": {
        "date": {"start": "2024-01-01", "end": "2024-02-15"},
        "category": "Planificación",
    },
    "Fase 2": {
        "date": {"start": "2024-02-10", "end": "2024-04-30"},
        "category": "Desarrollo",
    },
    "Fase 3": {
        "date": {"start": "2024-04-15", "end": "2024-06-01"},
        "category": "Testing",
    },
}

if __name__ == "__main__":
    # Create different timeline plots

    # Plotnine version (recommended for Quarto)
    plot1 = create_timeline_plotnine(timeline_data)
    print("Timeline con plotnine creado!")

    # Matplotlib version
    fig2 = create_timeline_matplotlib(timeline_data)
    print("Timeline con matplotlib creado!")

    # Seaborn version
    fig3 = create_timeline_seaborn(timeline_data)
    print("Timeline con seaborn creado!")

    # Gantt version
    plot4 = create_gantt_timeline(gantt_data)
    print("Timeline Gantt creado!")
```
