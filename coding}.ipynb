```python
df_falla = pd.read_csv(r"C:\Users\andmn\FALLA.csv")
# Convertir FECHA1 a datetime y crear columna año-mes
df_falla["FECHA1"] = pd.to_datetime(df_falla["FECHA1"])
df_falla["year_month"] = df_falla["FECHA1"].dt.to_period("M").astype(str)
df_falla = df_falla.query("FECHA1>='2024-08'")
# Obtener lista completa de meses y descripciones
all_months_falla = sorted(df_falla["year_month"].unique())
all_descriptions = sorted(df_falla["DESCRIPCION"].unique())

# Crear índice completo
full_index_falla = pd.MultiIndex.from_product(
    [all_months_falla, all_descriptions], names=["year_month", "DESCRIPCION"]
)
df_falla_full = (
    df_falla.groupby(["year_month", "DESCRIPCION"])
    .size()
    .reindex(full_index_falla, fill_value=0)
    .reset_index(name="count")
    .query("count!=0")
)

# # Pivot para gráfico
# pivot_falla = df_falla_full.pivot(
#     index="year_month", columns="DESCRIPCION", values="count"
# ).fillna(0)

df_falla_full  # .query("DESCRIPCION")
```
